## Latent Dirichlet Allocation (LDA)

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
# TF-IDF Vectorization
tfidf = TfidfVectorizer(min_df=2, max_df=0.95, stop_words='english')
tfid_chat = tfidf.fit_transform(data['chat'])
```

```python
# LDA Model
lda_model = LatentDirichletAllocation(n_components=5, random_state=42)
lda_model.fit(tfid_chat)
```

```python
for index,topics in enumerate(lda_model.components_):
    results=[tfidf.get_feature_names_out()[i]  for i in topics.argsort()]
    print(f'Index: {index}')
    print()
    print(f'Topics: {results}')
    print()
```

```python
topic_result = lda_model.transform(tfid_chat)
data['LDA_Topics'] = topic_result.argmax(axis=1)
```
